```python
import pandas as pd
import os

df = pd.read_csv("../data/bank.csv")
```

```python
os.makedirs(
    "../streaming_data/batches",
    exist_ok=True
)
```

```python
batch_size = 500

for i in range(0, len(df), batch_size):

    batch = df.iloc[i:i+batch_size]

    batch.to_csv(
        f"../streaming_data/batches/batch_{i}.csv",
        index=False
    )

print("Batch files created")
```
